```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
def dctree_using_gini(X_train,X_test,y_train):
    clf_gini=DecisionTreeClassifier(criterion="gini",random_state=100,max_depth=10,min_samples_leaf=15)
    clf_gini.fit(X_train,y_train)
    return clf_gini
```

```python
def dctree_using_entropy(X_train,X_test,y_train):
    clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=15,min_samples_leaf=10)
    clf_entropy.fit(X_train,y_train)
    return clf_entropy
```

```python
def predict(X_test,clf):
    ypred=clf.predict(X_test)
    return ypred;    
```
