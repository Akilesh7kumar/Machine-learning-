```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12.0,9.0)
#Preprocessing input
data = pd.read_csv("data.csv")
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.scatter(x,y)
plt.show()
```

```python
#building the model
m=0
c=0
L=0.0001
epochs = 1000
n = float(len(x))

#performing gradient descent
for i in range(epochs):
    y_pred = m*x+c
    d_m = (-2/n)*sum(x*(y-y_pred))
    d_c = (-2/n)*sum(y-y_pred)
    m = m- L*d_m
    c=c-L*d_c
    print(m,c,d_m)
```

```python
#Making prediction
y_pred = m*x+c
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)],color='red')
plt.show
```
